```python
import requests

def fetch_weather_rewa_weatherapi(api_key):
    city = "Rewa"
    url = f"http://api.weatherapi.com/v1/current.json?key={api_key}&q={city}&aqi=no"
    
    response = requests.get(url)
    response.raise_for_status()
    data = response.json()
    
    # Parse the relevant info
    weather_data = {
        "Location": data['location']['name'],
        "Region": data['location']['region'],
        "Country": data['location']['country'],
        "Local Time": data['location']['localtime'],
        "Temperature (C)": data['current']['temp_c'],
        "Condition": data['current']['condition']['text'],
        "Humidity (%)": data['current']['humidity'],
        "Wind (kph)": data['current']['wind_kph'],
        "Feels Like (C)": data['current']['feelslike_c'],
    }
    
    return weather_data

# --- Example Usage ---
API_KEY = "2434f2fa3ddf4180ad1223244250607"  # Replace with your actual WeatherAPI key
weather_info = fetch_weather_rewa_weatherapi(API_KEY)

# Pretty print
for k, v in weather_info.items():
    print(f"{k}: {v}")
```

```python
import requests

def fetch_5day_weather_forecast(api_key, city="Rewa", days=3):
    url = f"http://api.weatherapi.com/v1/forecast.json?key={api_key}&q={city}&days={days}&aqi=no&alerts=no"
    response = requests.get(url)
    response.raise_for_status()
    data = response.json()

    forecast_data = []
    for day in data['forecast']['forecastday']:
        forecast_data.append({
            "Date": day['date'],
            "Condition": day['day']['condition']['text'],
            "Avg Temp (°C)": day['day']['avgtemp_c'],
            "Max Temp (°C)": day['day']['maxtemp_c'],
            "Min Temp (°C)": day['day']['mintemp_c'],
            "Rainfall (mm)": day['day']['totalprecip_mm'],
            "Humidity (%)": day['day']['avghumidity']
        })

    return forecast_data

# Example Usage
API_KEY = "2434f2fa3ddf4180ad1223244250607"
forecast_info = fetch_5day_weather_forecast(API_KEY, city="Rewa", days=3)

# Pretty print
for day in forecast_info:
    print("\n".join([f"{k}: {v}" for k, v in day.items()]))
    print("-" * 40)
```

```python
import json

# Save to JSON
with open("weather_rewa.json", "w") as f:
    json.dump(forecast_info, f, indent=2)
```
